```python
import os
import shutil
import notebookutils # Import the utility

# --- 1. Define Paths using the official utility ---

# The location where you successfully read the file (Read-Only Source)
KAGGLE_CONFIG_DIR_SOURCE = f"{notebookutils.nbResPath}/builtin"
SOURCE_FILE_PATH = f"{KAGGLE_CONFIG_DIR_SOURCE}/kaggle.json"

# The ONLY guaranteed Writable temporary location (Target)
KAGGLE_CONFIG_DIR_SAFE = 'chmod 600 /tmp/.kaggle/kaggle.json' 
TARGET_FILE_PATH = f'{KAGGLE_CONFIG_DIR_SAFE}/kaggle.json'
```

```python
# --- 2. Copy the file ---

# Create the target directory
os.makedirs(KAGGLE_CONFIG_DIR_SAFE, exist_ok=True)

try:
    # Use standard Python copy, which is faster for local cluster paths
    shutil.copyfile(SOURCE_FILE_PATH, TARGET_FILE_PATH)
    print("‚úÖ kaggle.json successfully copied to a writable location.")

except FileNotFoundError:
    print(f"‚ùå ERROR: Source file not found at {SOURCE_FILE_PATH}. Check file name.")
except Exception as e:
    print(f"‚ùå ERROR during copy: {e}")
```

```python
# --- 3. Set the Environment Variable ---

# Set the KAGGLE_CONFIG_DIR environment variable to the *new, safe directory*
os.environ['KAGGLE_CONFIG_DIR'] = KAGGLE_CONFIG_DIR_SAFE
print(f"KAGGLE_CONFIG_DIR set to: {os.environ.get('KAGGLE_CONFIG_DIR')}")
```

```python
try:
    from kaggle.api.kaggle_api_extended import KaggleApi
    api = KaggleApi()
    api.authenticate()
    
    print("\n\n‚ú® FINAL SUCCESS: Kaggle API authenticated successfully! üéâ")

except Exception as e:
    print(f"\n\n‚ùå FINAL FAILURE: Authentication failed.")
    print("If you see this error, please re-download a fresh kaggle.json file, re-upload it, and try again.")
    print(f"Error: {e}")
```
